```python
# This function is used to convert List of Strings to Simple List
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def  convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def  convert3(obj):
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def  fetch_director(obj):
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies ['crew'] = movies['crew'].apply(fetch_director)
```

```python
# I will convert this movies['overview'] value in the LIST
# Format so then i can concatenate this with other values
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
# Now i have to convert this 'Sam Worthington'
# into like this 'SamWorthington' to overcome the problem like similar names
# like 'Sam Mendes'
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# Now we will concatenate all these columns 
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
stem("'in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron'")
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# now we will convert these tags into vectors 
# I will convert it into (x,y) form and plot onto 2-D space
# now 5000 movies will be on 2-D space and now we will recommend other 
# person movies based on the closest ones that is near to that vector

# text -> vectors = this is called TEXT VECTORIZATION
# we have 3 techniques here: 
# 1. Bag of words (we will use this in this project) (easy technique)
# 2. Lfidf
# 3. word2vec
```

```python
# Bag of words
# Take example
# In this technique we will combine all tags
# assume 1 tag = 100 words
# so 5000 tag = 5 lakh word
# from 5 lakh word we will find top 5000 words that occurs frequently 

# now we will compare these 5000 words from particular movie 100 tag
# and made a table(5000,5000) where 5000 movies will be aligned in rows
# and 5000 frequently words will be aligned in columns
# and we will mark how much time each word comes in 100 word tag of that particular movie

# Table like this: 

#           w1 w2 w3 _ _ _ _  w5000 (words frequently used from 5 lakh)
# m1        5  3  0           1
# m2        2  0  1           3 
# m2        1  1  1           1
# _
# m5000
# (movies)

# Here Row m1 is called vector
# So we have 5000 vectors here
# Now arrange these rows in D-space and this is called vectors

# Que: why i choose 5000 words only?
# Sol: this will depend totally on you but not less or not more is generally considered


# One important thing we will do here is that we will not consider STOP WORDS HERE (ARE AND OF TO FROM)
# STOP WORDS USED FOR SENTENCE FORMATION BUT NO CONTRIBUTION IN DEFINING SENTENCE MEANING 

```

```python
 from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# we have 4806 vectors 
# so we have to calculate distance from each movies to the other

# here we will not calculate the eucleidian distance instead we will calculate the
# "Cosine distance (angle between them)"

# for this we will use the cosine similarity function
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
# Every movie distance with other movie 
```

```python
similarity = cosine_similarity(vectors)
```

```python
# now we want only first 5 recommended movie so except first one
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x: x[1])[1:6]
```

```python
# if we use sorting then we will forgot the index becuase now it is sorted
# so we have to make sure that we will know index of that movie also
# then only we can say 1st movie ka 2nd ke sath kitna
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```
